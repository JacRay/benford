```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math
import cv2
```

```python
def firstDigit(number):
    digits = (int)(math.log10(number)) 
    number = (int)(number / pow(10, digits)) 
    return number
def graph(image_name,flag):
    
    img = cv2.imread(image_name,0) #load image
    norm_img = np.zeros((800,800))  #create a matrix 800*800 with all value zero
    final_img = cv2.normalize(img,  norm_img, 0, 255, cv2.NORM_MINMAX) # normlize image clear its noise
    cv2.imwrite('city_normalized.jpg', final_img) #write image to new file
    img = cv2.imread('./city_normalized.jpg',0) # read the normalized image
    imf = np.float32(img)/255.0 #changing datatype
    dst = cv2.dct(imf)
    img = np.uint8(dst)*255.0 #unsigned integer datatype changing
    
    col = dst.reshape(dst.shape[0]*dst.shape[1],).tolist() #reshape to row and col size
    length = len(col)
    
    #initializing count of each digit
    dict_freq = {}
    for digit in range(1,10):
        dict_freq[str(digit)] = 0

    #finding no of occurrence of first digits
    for value in col:
        if (type(value) is int or type(value) is float) and value>0:
    #         print(value)
            f_digit = firstDigit(value*1000000000000000000000000)
            dict_freq[str(f_digit)]+=1

    # initializing count according to benfords law
    benford_dict = {}
    for digit in range(1,10):
        benford_dict[str(digit)] = math.log10(1+1/digit)
    freq_benf = np.array(list(benford_dict.values()))
    total_entries = 0
    #finding total entries
    for i in list(dict_freq.values()):
        total_entries+=i
    digits = list(dict_freq.keys())
    
    digits = list(dict_freq.keys())
    frequency = list(dict_freq.values())
    #finding total no of required distribution
    benford_freq = freq_benf * total_entries
    #graph with x range dict_freq
    plt.bar(range(len(dict_freq)),frequency,tick_label=digits,color = 'silver')
    if flag==0:
        plt.plot(digits,frequency,label="Original Image Analysis",color='black',linestyle='dashed',linewidth=3,marker='o')
    else:
        plt.plot(digits,frequency,label="Modified Image Analysis",color='black',linestyle='dashed',linewidth=3,marker='o')
    plt.plot(digits,benford_freq,label="Actual Benford's Analysis",color='blue',linestyle='dashed',linewidth=3,marker='*')
    plt.xlabel('Digits')
    plt.ylabel('Frequency')
    plt.legend(loc='best')
    plt.savefig('result.png', dpi=300, bbox_inches='tight')
    plt.show()
```

```python
graph('./original.jpg',0)
```

```python
graph('./modified.jpg',1)
```
